```python
import pandas as pd
import numpy as np
```

# Offenses in each City

```python
offense = pd.DataFrame(pd.read_excel("Data/Table_8_Offenses_Known_to_Law_Enforcement_by_State_by_City_2021.xlsx"))
offense
```

```python
offense = offense.drop([0, 1, 2, 5878])
offense
```

```python
offense = offense.rename(columns={"Table 8": "State",
                          "Unnamed: 1": "City",
                          "Unnamed: 2": "Population",
                          "Unnamed: 3": "Violent_crime",
                          "Unnamed: 4": "Murder_and_nonnegligent_manslaughter",
                          "Unnamed: 5": "Rape",
                          "Unnamed: 6": "Robbery",
                          "Unnamed: 7": "Aggravated_assault",
                          "Unnamed: 8": "Property_crime",
                          "Unnamed: 9": "Burglary",
                          "Unnamed: 10": "Larceny-theft",
                          "Unnamed: 11": "Motor_vehicle_theft",
                          "Unnamed: 12": "Arson"})
offense
```

```python
offense = offense.reset_index(drop=True)
offense
```

```python
state = ""
for index, row in offense.iterrows():
    if pd.notnull(row['State']):
        state = row['State']
    else:
        row['State'] = state
        
offense
```

```python
offense = offense.astype({'State':'string', 'City':'string', 'Population':'float', 'Violent_crime':'float', 'Murder_and_nonnegligent_manslaughter':'float', 'Rape':'float', 'Robbery':'float', 'Aggravated_assault':'float', 'Property_crime':'float', 'Burglary':'float', 'Larceny-theft':'float', 'Motor_vehicle_theft':'float', 'Arson':'float'})
offense.dtypes
```

# Police in each City

```python
police = pd.DataFrame(pd.read_excel("Data/Table_78_Full_time_Law_Enforcement_Employees_by_State_by_Cities_2021.xlsx"))
police
```

```python
police = police.drop([0, 1, 2, 8656])
police
```

```python
police = police.rename(columns={"Table 78": "State",
                          "Unnamed: 1": "City",
                          "Unnamed: 2": "Population",
                          "Unnamed: 3": "Total_law_enforcement_employees",
                          "Unnamed: 4": "Total_officers",
                          "Unnamed: 5": "Total_civilians"})
police
```

```python
police = police.reset_index(drop=True)
police
```

```python
state = ""
for index, row in police.iterrows():
    if pd.notnull(row['State']):
        state = row['State']
    else:
        row['State'] = state

police
```

```python
police = police.astype({'State':'string', 'City':'string', 'Population':'float', 'Total_law_enforcement_employees':'int', 'Total_officers':'int', 'Total_civilians':'int'})
police.dtypes
```

# Combine Tables

```python
police_offense = pd.merge(offense, police, how='inner', on=['State', 'City'])
police_offense
```

```python
police_offense = police_offense.drop(['Murder_and_nonnegligent_manslaughter', 'Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny-theft', 'Motor_vehicle_theft', 'Arson', 'Population_y', 'Total_officers', 'Total_civilians'], axis=1)
police_offense
```

```python
police_offense['Total_crime'] = police_offense['Violent_crime'] + police_offense['Property_crime']
police_offense
```

```python
police_offense = police_offense.drop(['Violent_crime', 'Property_crime'], axis=1)
police_offense = police_offense.rename(columns={"Population_x": "Population"})
police_offense
```

```python
police_offense['employees_perCapita'] = police_offense["Total_law_enforcement_employees"] / police_offense["Population"]
police_offense['crime_perCapita'] = police_offense["Total_crime"] / police_offense["Population"]
police_offense
```

# Remove Outliers and Apply Regression 

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
```

```python
x = np.array(police_offense["employees_perCapita"])
y = np.array(police_offense["crime_perCapita"])

plt.scatter(x, y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x.astype(float), y.astype(float))
print(slope)
def myfunc(x):
  return slope * x + intercept       # y = mx + c

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
