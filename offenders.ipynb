# Adult and Juvenile Age Category

```python
AJ1 = pd.DataFrame(pd.read_excel("Data/offenders/Offenders_Adult_and_Juvenile_Age_Category_by_Offense_Category_2021.xlsx"))
AJ1
```

```python
AJ1 = AJ1.drop([0, 1, 2, 3, 4, 32, 33])
AJ1
```

```python
AJ1 = AJ1.rename(columns={"Offenders": "Offense_Category",
                          "Unnamed: 1": "Total_Offenders",
                          "Unnamed: 2": "Adult",
                          "Unnamed: 3": "Juvenile",
                          "Unnamed: 4": "Unkown Age",})
AJ1
```

```python
AJ2 = AJ1[(AJ1["Offense_Category"] == "Crimes Against Persons") |
          (AJ1["Offense_Category"] == "Crimes Against Property") |
          (AJ1["Offense_Category"] == "Crimes Against Society") |
          (AJ1["Offense_Category"] == "Total")]
            
AJ2
```

```python
AJ1 = AJ1[((AJ1.Offense_Category != "Crimes Against Persons") &
          (AJ1.Offense_Category != "Crimes Against Property") &
          (AJ1.Offense_Category != "Crimes Against Society"))]
AJ1
```

```python
AJ1 = AJ1.reset_index(drop=True)
AJ2 = AJ2.reset_index(drop=True)
display(AJ1)
display(AJ2)
```

```python
AJ1["Crimes Against"] = [np.nan,
                         "Persons", "Persons", "Persons", "Persons", "Persons",
                        "Property", "Property", "Property", "Property", "Property", "Property", "Property", "Property", "Property", "Property", "Property", "Property",
                        "Society", "Society", "Society", "Society", "Society", "Society"]
AJ1
```
